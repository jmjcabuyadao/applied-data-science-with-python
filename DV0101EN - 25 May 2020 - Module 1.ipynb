```python
canada_df = pd.read_excel('canada_immigration.xlsx', sheet_name='Canada by Citizenship',
                         skiprows=range(20), skipfooter=2)
canada_df.head()
```

```python
canada_by_country = pd.read_excel('canada_immigration.xlsx', sheet_name='Canada by Citizenship',
                                 skiprows=range(20), skipfooter=2)
canada_by_country.set_index('OdName', inplace=True)
canada_by_country.index.names = ['Country']
canada_by_country.head()
```

```python
years = list(map(int, range(1980, 2014)))
canada_by_country['Total'] = canada_by_country[years].sum(axis=1)
canada_by_country
```

```python
canada_by_country.loc['Haiti', years].plot(kind='line')
plt.title('Immigration From Haiti 1980-2013')
plt.xlabel('Year')
plt.ylabel('No. of immigrants')

plt.show()
```

We can easily add more countries to line plot to make meaningful comparisons immigration from different countries. 

**Question:** Let's compare the number of immigrants from India and China from 1980 to 2013.

Step 1: Get the data set for China and India, and display dataframe.

```python
### type your answer here
df_CI = canada_by_country.loc[['China', 'India'], years]
df_CI
```

Step 2: Plot graph. We will explicitly specify line plot by passing in `kind` parameter to `plot()`.

```python
### type your answer here
china = df_CI.loc['China', years]
india = df_CI.loc['India', years]
china.plot(kind='line')
india.plot(kind='line')

plt.title('Immigration from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
df_CI = df_CI.transpose()
df_CI.head()

df_CI.index = df_CI.index.map(int) # let's change the index values of df_CI to type integer for plotting
df_CI.plot(kind='line')

plt.title('Immigrants from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

**Question:** Compare the trend of top 5 countries that contributed the most to immigration to Canada.

```python
### type your answer here
canada_by_country.sort_values(by='Total', ascending=False, axis=0, inplace=True)
top_5_immig = canada_by_country.head(5)
top_5_immig = top_5_immig.loc[:, years]

top_5_immig = top_5_immig.transpose()
top_5_immig.index = df_CI.index.map(int) # let's change the index values of df_CI to type integer for plotting
top_5_immig.plot(kind='line', figsize=(14, 8))

plt.title('Top 5 Immigration to Canada')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```
