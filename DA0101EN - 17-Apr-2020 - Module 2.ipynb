```python
import pandas as pd
import numpy as np
```

```python
headers = [
    'symboling',
    'normalized-losses',
    'make',
    'fuel-type',
    'aspiration',
    'num-of-doors',
    'body-style',
    'drive-wheels',
    'engine-location',
    'wheel-base',
    'length',
    'width',
    'height',
    'curb-weight',
    'engine-type',
    'num-of-cylinders',
    'engine-size',
    'fuel-system',
    'bore',
    'stroke',
    'compression-ratio',
    'horsepower',
    'peak-rpm',
    'city-mpg',
    'highway-mpg',
    'price'
]
autos = pd.read_csv('auto-dataset.csv', names=headers)
autos
```

```python
autos['normalized-losses'].replace('?', np.nan, inplace=True)
autos
```

```python
# won't be able to compute mean because of NaN treated as string, need to cast values
autos['normalized-losses'] = pd.to_numeric(autos['normalized-losses'], errors = 'coerce')
normalized_losses_mean = autos['normalized-losses'].mean(skipna=True)
autos['normalized-losses'].replace(np.nan, normalized_losses_mean, inplace=True)
```

```python
autos
```

```python
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 300)
autos
```

```python
city_lp100km = round(235/autos['city-mpg'], 1)
# adding new column on specific position, zero-index
autos.insert(24, 'city-lp100km', city_lp100km)
autos
```

```python
len_min_max = (autos['length'] - autos['length'].min())/(autos['length'].max()-autos['length'].min())
wid_min_max = (autos['width'] - autos['width'].min())/(autos['width'].max()-autos['width'].min())
hei_min_max = (autos['height'] - autos['height'].min())/(autos['height'].max()-autos['height'].min())

normalized_autos = pd.DataFrame([])
normalized_autos.insert(0, 'norm-length', round(len_min_max, 3))
normalized_autos.insert(1, 'norm-width', round(wid_min_max, 3))
normalized_autos.insert(2, 'norm-height', round(hei_min_max, 3))
normalized_autos
```

```python
fuel_type = pd.get_dummies(autos['fuel-type'])

# concatenate all data frames to create new one
autos = pd.concat([
    autos[[
        'symboling',
        'normalized-losses',
        'make',
        'fuel-type'
    ]], 
    fuel_type,
    autos[[
        'aspiration',
        'num-of-doors',
        'body-style',
        'drive-wheels',
        'engine-location',
        'wheel-base',
        'length',
        'width',
        'height'
    ]], 
    normalized_autos, 
    autos[[
        'curb-weight',
        'engine-type',
        'num-of-cylinders',
        'engine-size',
        'fuel-system',
        'bore',
        'stroke',
        'compression-ratio',
        'horsepower',
        'peak-rpm',
        'city-mpg',
        'highway-mpg',
        'price'
    ]]
    ], axis=1)
autos
```
