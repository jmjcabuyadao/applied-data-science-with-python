```python
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline 
```

```python
autos['price'].replace('?', np.nan, inplace=True)
autos['bore'].replace('?', autos['bore'].mode(), inplace=True)
autos['stroke'].replace('?', autos['stroke'].mode(), inplace=True)
autos['num-of-doors'].replace('?', autos['num-of-doors'].mode(), inplace=True)
autos['price'] = autos['price'].astype(float)
autos.dropna(subset=['price'], inplace=True)
```

```python
drive_body_price = autos[['drive-wheels', 'body-style', 'price']]
drp_group = drive_body_price.groupby(['drive-wheels', 'body-style'], as_index=False).mean()
drp_group
```

```python
drp_pivot = drp_group.pivot(index='drive-wheels', columns='body-style')
drp_pivot
```

```python
plt.pcolor(drp_pivot, cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
make_price = autos[['make', 'price']]
make_price_group = make_price.groupby(['make'], as_index=False).mean()
make_price_group
```

```python
make_price = autos[['make', 'price']]
make_price_group = make_price.groupby(['make'], as_index=False)
make_anova = stats.f_oneway(make_price_group.get_group('honda')['price'], make_price_group.get_group('subaru')['price'])
make_anova
```

```python
autos['peak-rpm'].replace('?', np.nan, inplace=True)
autos['peak-rpm'] = pd.to_numeric(autos['peak-rpm'], errors = 'coerce')
peak_rpm_mean = autos['peak-rpm'].mean(skipna=True)
autos['peak-rpm'].replace(np.nan, peak_rpm_mean, inplace=True)
autos['peak-rpm'] = autos['peak-rpm'].astype(int)
sns.regplot(x='peak-rpm', y='price', data=autos)
plt.ylim(0,)
```

```python
# coeff, p_val = stats.pearsonr(autos['horsepower'], autos['price'])
autos['horsepower'].replace('?', np.nan, inplace=True)
autos['horsepower'] = pd.to_numeric(autos['horsepower'], errors = 'coerce')
autos['horsepower'].replace(np.nan, autos['horsepower'].mean(skipna=True), inplace=True)
autos['horsepower'] = autos['horsepower'].astype(int)
```

```python
coeff, p_val = stats.pearsonr(autos['horsepower'], autos['price'])
print(coeff)
print(p_val)
```
